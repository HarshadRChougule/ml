```python
#train the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df, y_df, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
```

```python
X_train.shape
```

```python
LR.fit(X_train,y_train)
```

```python
LR.score(X_train,y_train)*100
```

```python
LR.score(X_test,y_test)*100
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df, y_df, test_size=0.30, random_state=30)
```

```python
print((LR.score(X_train,y_train)*100),(LR.score(X_test,y_test)*100))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
y_pred = LR.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test,y_pred),fmt="d",xticklabels=["Belign","Malignant"],yticklabels=["Belign","Malignant"],cmap="Reds",annot=True)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score, accuracy_score
```

```python
print('precision_score ', precision_score(y_test,y_pred), ' recall_score',recall_score(y_test,y_pred), ' f1_score',f1_score(y_test,y_pred),' accuracy_score', accuracy_score(y_test,y_pred))
```
